```python
# write functions to reuse code
def clean_price_col(df):
   df['price'] = df['price'].str.replace('$', '').str.replace(',', '')
   df['price'] = df['price'].astype('float64')
   return df

def merge_reviews_and_listings_csvs(path_to_reviews, path_to_listings):
    # first read in files
    listing_df = pd.read_csv(path_to_listings)
    reviews_df = pd.read_csv(path_to_reviews)
    # filter out unwanted cols
    listing_df = listing_df.loc[:, ['id', 'host_id', 'amenities', 'price', 'review_scores_rating', 'review_scores_accuracy',
       'review_scores_cleanliness', 'review_scores_checkin',
       'review_scores_communication', 'review_scores_location',
       'review_scores_value', 'reviews_per_month', 'number_of_reviews']]
    # rename 'id' col to be 'listing_id'
    listing_df = listing_df.rename(columns={'id': 'listing_id'})

    merged_df = pd.merge(reviews_df, listing_df, on="listing_id", how='inner')
    merged_df.dropna(inplace=True)

    return clean_price_col(merged_df)
```

```python
def get_sample(df):
    # filter reviews from listings w/ median or 1 std dev away from median
    median = df['number_of_reviews'].median()
    std_dev = df['number_of_reviews'].std()
    filter =  (df['number_of_reviews'] >= median-std_dev) & (df['number_of_reviews'] <= median+std_dev )
    # print("median: ", median)
    # print("std dev: ", std_dev)
    # print(filter)
    filtered_df = df.loc[filter, :]
    #filtered_df.info()

    # randomly sample 600 observations
    sampled_df = filtered_df.sample(n=600, random_state=22)

    return sampled_df
```

```python
san_diego_sample = get_sample(san_diego)
```
